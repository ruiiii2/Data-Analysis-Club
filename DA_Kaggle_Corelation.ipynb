```python
import pandas as pd
import matplotlib.pyplot as plt

data_df = pd.read_csv("./US_Accidents_Dec21_updated.csv")
data_df=data_df.dropna()
```

```python
df = pd.DataFrame(data_df)
```

```python
df.info()
```

```python
df["Start_Time"] = pd.to_datetime(df["Start_Time"])
```

```python
df['year'] = df["Start_Time"].dt.year
```

```python
drop_list = list(df['Weather_Condition'].value_counts().tail(65).index)
```

```python
for i in drop_list:
    df.drop(df[df["Weather_Condition"]==i].index,inplace=True)
```

```python
df['Weather_Condition'].value_counts()
```

```python
weather_list_float = ['Temperature(F)','Wind_Chill(F)','Humidity(%)','Pressure(in)','Visibility(mi)','Wind_Speed(mph)','Precipitation(in)']
for i in weather_list_float:
    df.boxplot(i,by="Weather_Condition",rot=90,figsize=(10,6))
```

```python
weather_list_float = ['Temperature(F)','Wind_Chill(F)','Humidity(%)','Pressure(in)','Visibility(mi)','Wind_Speed(mph)','Precipitation(in)']
for i in weather_list_float:
    df.boxplot(i,by="Severity",rot=45,figsize=(7,6))
```
